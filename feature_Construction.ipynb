```python
import pandas as pd
import numpy as np

import seaborn as sns

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('/content/sample_data/tested (2).csv')[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.sample(4)
```

```python
df.dropna(inplace=True)
```

```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
X.sample(5)
```

```python
y.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=25))
```

# Apply Feature Construction

```python
X['Family_size']= X['SibSp'] + X['Parch'] + 1
```

```python
X.head()
```

```python
def myfunc(num):
  if num == 1:
    #alone
    return 0
  elif num >1 and num <=4:
    # small family
    return 1
  else:
    # large family
    return 2
```

```python
myfunc(4)
```

```python
X['Family_type'] = X['Family_size'].apply(myfunc)
```

```python
X.sample(5)
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=5))
```
