```python
# 📊 Cross Validation (CV)

# Cross Validation is used to check
# how well the model performs on unseen data
# In cv=5:
# Dataset splits into 5 parts (folds)
# Step:
# Train on 4 folds
# Test on 1 fold
# Repeat 5 times
# Final result:
# Average performance of all folds
# 🎯 Helps reduce overfitting
# and gives more reliable accuracy

# 📦 GridSearchCV
# GridSearchCV is used for Hyperparameter Tuning
# It:
# ✅ Tries ALL parameter combinations
# ✅ Uses Cross Validation internally
# ✅ Finds best parameters automatically
# Example:
# C = [1,10]
# kernel = ['linear','rbf']
# It tests:
# (1, linear)
# (1, rbf)
# (10, linear)
# (10, rbf)
# ⚠️ Accurate but slower

# 📦 RandomizedSearchCV
# RandomizedSearchCV is also used
# for Hyperparameter Tuning
# Instead of testing ALL combinations,
# it tests RANDOM combinations
# n_iter = number of random combinations tested
# ⚡ Faster than GridSearchCV
# especially for large parameter spaces
# 🎯 Useful when dataset or parameter list is large
```

```python
df = pd.read_csv('/kaggle/input/datasets/yasserh/breast-cancer-dataset/breast-cancer.csv')
```

```python
X = df.iloc[:,1:]
X
```

```python
y = df.iloc[:,0]
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
# Split the data (default test_size is 0.25)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
model = LogisticRegression(penalty='l2', solver='liblinear')
fs = StandardScaler()
X_train_scaled = fs.fit_transform(X_train)
X_test_scaled = fs.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
model.fit(X_train_scaled,y_train)
y_pred = model.predict(X_test_scaled)
score = accuracy_score(y_test,y_pred )
print(f'The score of this model is {score*100}%')
```

```python
#cross Validation cv=5 means 5 fold 
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X_train_scaled, y_train, cv=5)

print(f"Scores for each fold: {scores}")
print(f"Mean Accuracy: {scores.mean():.2f}")

  # 🔄 5-Fold Cross Validation
# Dataset splits into 5 parts
# Step:
# Train on 4 parts
# Test on 1 part
# Repeat 5 times
# Final result:
# 5 accuracy scores
```

```python
from scipy.stats import randint
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

dt_clf = DecisionTreeClassifier()

# 📦 Dictionary containing hyperparameters
# used for RandomizedSearchCV

param_dist = {
    # 📊 criterion → split quality measure
    # 'gini' → Gini Index
    # 'entropy' → Information Gain
    "criterion": ["gini", "entropy"],
    # 🌲 max_depth → maximum depth of tree
    # randint(1, 20)
    # → Random value between 1 and 20
    "max_depth": randint(1, 20),
    # 🍃 min_samples_leaf
    # → Minimum samples required in a leaf node
    # randint(1, 10)
    # → Random value between 1 and 10
    "min_samples_leaf": randint(1, 10),
    # 📦 max_features
    # → Number of features considered at each split
    # X.shape[1]
    # → Total number of columns/features
    # randint(1, X.shape[1])
    # → Random number of features selected
    "max_features": randint(1, X.shape[1])
}

# 📦 Create RandomizedSearchCV object
# Used for Hyperparameter Tuning
random_search = RandomizedSearchCV(
    dt_clf,
    # 🌳 Decision Tree model to train
    param_dist,
    # 📊 Dictionary containing hyperparameters
    # and their possible values
    n_iter=20,
    # 🔁 Number of random parameter combinations to try
    # Instead of checking ALL combinations,
    # RandomizedSearchCV tests only 20 random combinations
    cv=5,
    # 🔄 5-Fold Cross Validation
    # Dataset splits into 5 parts
    # Train on 4 parts
    # Test on 1 part
    # Repeat 5 times

    random_state=42
    # 🔁 Ensures same random results every run
)

random_search.fit(X_train_scaled, y_train)
# 6. Display all scores
results = pd.DataFrame(random_search.cv_results_)
results
```

```python
# 📦 Import SVM algorithm
# svm → Support Vector Machine module
from sklearn import svm
# 📦 Import Random Forest Classifier
# Ensemble learning algorithm using multiple decision trees
from sklearn.ensemble import RandomForestClassifier
# 📦 Import Logistic Regression model
# Used for classification problems
from sklearn.linear_model import LogisticRegression
# 📦 Import GridSearchCV
# Used for Hyperparameter Tuning

# GridSearchCV automatically:
# ✅ Tries different parameter combinations
# ✅ Finds best model parameters
# ✅ Uses Cross Validation
from sklearn.model_selection import GridSearchCV
# 📦 Import pandas for DataFrame creation
import pandas as pd
# 📊 Dictionary containing:
# → Models
# → Their hyperparameters

model_params = {

    # 🌳 SVM Model Section
    'svm': {

        # 🧠 Create SVM model
        'model': svm.SVC(gamma='auto'),

        # 📦 Hyperparameters to test
        'params': {

            # 🎛 C → Regularization parameter
            # Small C → wider margin
            # Large C → less tolerance for mistakes

            'C': [1, 10, 20],

            # 🧩 kernel → Type of decision boundary
            # 'linear' → straight line
            # 'rbf' → curved boundary

            'kernel': ['rbf', 'linear']
        }
    },

    # 🌲 Random Forest Section
    'random_forest': {
        # 🧠 Create Random Forest model
        'model': RandomForestClassifier(),
        # 📦 Hyperparameters
        'params': {
            # 🌳 n_estimators
            # Number of decision trees in forest

            'n_estimators': [1, 5, 10]
        }
    },

    # 📈 Logistic Regression Section
    'logistic_regression': {

        # 🧠 Create Logistic Regression model
        'model': LogisticRegression(

            # ⚙️ solver → optimization algorithm
            # 'liblinear' works well for small datasets

            solver='liblinear'
        ),

        # 📦 Hyperparameters
        'params': {

            # 🎛 C → Regularization strength
            # Smaller C → stronger regularization

            'C': [1, 5, 10]
        }
    }
}


# 📋 Empty list to store model results
scores = []


# 🔁 Loop through each model in dictionary

for model_name, mp in model_params.items(): #model_name and mp is (key,value)
    #key = model_name =svm,Logistic_regression
    #value = mp like solver is which 

    
    # 🧠 Create GridSearchCV object

    clf = GridSearchCV(

        mp['model'],
        # 🤖 Model to train

        mp['params'],
        # 📦 Hyperparameter combinations to test

        cv=5,
        # 🔁 5-Fold Cross Validation

        # Data splits into 5 parts
        # Train on 4 parts
        # Test on 1 part
        # Repeat 5 times

        return_train_score=False
        # 🚫 Do not return training score
        # Saves memory
    )


    # 🎯 Train model using all parameter combinations
    clf.fit(X_train_scaled, y_train)


    # 📊 Store results in list

    scores.append({

        # 🏷️ Model name
        'model': model_name,

        # 🥇 Best cross-validation score
        'best_score': clf.best_score_,

        # ⚙️ Best hyperparameters found
        'best_params': clf.best_params_
    })


# 📋 Convert results list into DataFrame
df = pd.DataFrame(scores)


# 👁️ Display final comparison table
df
```
