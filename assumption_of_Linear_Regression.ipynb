**What is Assumptin of the Linear Regression.**

Linear Regression works properly only when some assumptions(Guess) about the data are satisfied. These are called the assumptions of linear regression. Each assumption describes a condition that should be approximately true for the model to give reliable results.

**There are Five type of Assumption of the Linear Regression**

1️⃣ Linearity

2️⃣ Independence of Error(No autocorrelation among residuals(Errors))

3️⃣ Homoscedasticity

4️⃣ Normality of Errors(Residual)

5️⃣ No Multicollinearity (for multiple linear regression)


1️⃣ **Linearity**

Meaning:
The relationship between the input variables (X) and the output (Y) should be linear.

Interpretation:
If X increases, Y should change in a straight-line pattern.
Example:
Hours studied → Exam score.

If the relationship is curved instead of straight, linear regression may not fit well.

5️⃣ **No Multicollinearity (for multiple linear regression means 2 columns(input) or more than 2)**

In multiple linear regression, the independent variables (features) should not be strongly correlated with each other.

Example  — Education and Age

Predicting salary.

Features:
age,(X1)
years of experience(X2)

Older people usually have more experience.,
So these variables often move together.

The model cannot clearly determine which variable drives salary(Y).

Why this is a problem ⚠️
When two features carry similar information:

1️⃣ The model cannot clearly decide which variable is responsible for the prediction.

2️⃣ The coefficients become unstable.

3️⃣ Small changes in data can cause large changes in coefficients.


**Multicollinearity happens when:
    One feature can almost be predicted from another feature.**
    
https://media.geeksforgeeks.org/wp-content/uploads/20250807113336406829/understanding_multicollinearity-.webp

How to detect multicollinearity

Common methods:

1️⃣ Correlation matrix
If correlation between features > 0.8 or 0.9, it may be a problem.

2️⃣ Variance Inflation Factor (VIF)
Rule of thumb:

VIF	Meaning

1	No correlation

1–5	Moderate

10	Serious multicollinearity

How to fix it
Common approaches:

Remove one of the correlated features:-
Combine features,
Use PCA,
Use regularization (Ridge / Lasso regression)

4️⃣ **Normality of Errors(Resudual)**

Meaning:
The errors (residuals) of the model should follow a normal distribution (a bell-shaped curve).

Intuition

If a regression model is good, the errors should behave randomly around zero.

That means:

many small errors

fewer large errors

positive and negative errors balanced

When plotted, the errors should look like a bell-shaped distribution.

How to check it

Common methods:

1️⃣ Histogram of residuals
Should look like a bell curve.

2️⃣ Q–Q plot

3️⃣ Homoscedasticity

Meaning:
The variance(spread) of errors should remain constant for all values of X.

Interpretation:

The spread of residuals(errors) should be roughly equal everywhere.

Good case:
Residuals(errors) look like a random horizontal band.

Bad case (heteroscedasticity):

Residuals form a cone or funnel shape.

https://miro.medium.com/1*OnODBB93DWEr_sZoVjZS4A.png


2️⃣ Independence of Error(No autocorrelation among residuals(Errors))

Simple intuition

Errors should behave randomly, not follow a pattern.

Example of independent errors:

🔹 Observation 1 error = 2

🔹 Observation 2 error = −1

🔹 Observation 3 error = 0.3

🔹 Observation 4 error = −0.5

Each error is separate and random.
