```python
# 📊 Bagging Regressor

# BaggingRegressor is an Ensemble Learning algorithm
# used for Regression problems

# 🎯 Goal:
# Improve prediction accuracy
# and reduce overfitting

# 🧠 Idea:
# Train multiple regression models
# on different random subsets of data
# and combine their outputs

# 📦 Bootstrap Sampling:
# Random sampling WITH replacement

# Means:
# Same data point can appear multiple times
# in one sampled dataset

# 📈 Final Prediction:
# Average of predictions from all models


# 🔁 Working:

# Step 1️⃣
# Create multiple random datasets

# Step 2️⃣
# Train separate regressors on each dataset

# Step 3️⃣
# Each model predicts output

# Step 4️⃣
# Take average of all predictions


# 📦 Common Base Models:
# 🌳 DecisionTreeRegressor
# 📈 Linear Regression


# 🎯 Example:

# Model1 Prediction → 100
# Model2 Prediction → 110
# Model3 Prediction → 120

# Final Prediction:
# (100 + 110 + 120) / 3 = 110


# ✅ Advantages:
# Reduces overfitting
# Better stability
# Improves accuracy

# ❌ Disadvantages:
# Slower
# More memory usage
# Harder to interpret


# 📊 Simple Diagram:

#               📥 Original Dataset
#                        │
#      ┌─────────────────┼─────────────────┐
#      │                 │                 │
#   📦 Sample1        📦 Sample2        📦 Sample3
#      │                 │                 │
#   📈 Model1         📈 Model2         📈 Model3
#      │                 │                 │
#      └─────────────────┼─────────────────┘
#                        │
#                📊 Average Prediction
#                        │
#                 ✅ Final Output
```

```python
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=900, n_features=20 ,n_informative=10, random_state=42)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, test_size=0.2, random_state=42)
print('Train/Test Sets Sizes : ',X_train.shape, X_test.shape, y_train.shape, y_test. shape)
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
knn = KNeighborsRegressor()

lr.fit(X_train, y_train)
dt.fit(X_train,y_train)
knn.fit(X_train,y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = dt.predict(X_test)
y_pred3 = knn.predict(X_test)

print("R^2 score for LR",r2_score(y_test,y_pred1))
print("R^2 score for DT",r2_score(y_test,y_pred2))
print("R^2 score for KNN",r2_score(y_test,y_pred3))
```

So “Estimator=None” Does NOT Mean “No Model”
It means:
"Use sklearn's default model internally"

which is:

DecisionTreeClassifier() for classification
DecisionTreeRegressor() for regression 

```python
#So “Estimator=None” Does NOT Mean “No Model”
#It means:
#"Use sklearn's default model internally"
#which is:
#DecisionTreeClassifier() for classification
#DecisionTreeRegressor() for regression 

from sklearn.ensemble import BaggingRegressor

bag_regressor = BaggingRegressor(estimator= None , random_state=1)
bag_regressor.fit(X_train, y_train)
```

```python
y_preds = bag_regressor.predict(X_test)

print('Training Coefficient of R^2 :',bag_regressor.score(X_train, y_train))
print('Test Coefficient of R^2 :',bag_regressor.score(X_test, y_test))
print(f'The accuacacy score of this model is ',r2_score(y_test,y_preds))
```

# **Applying Bagging Regression with GridSearchCV**

```python
from sklearn.ensemble import BaggingRegressor

params = {'estimator': [None, LinearRegression(), KNeighborsRegressor()],
'n_estimators': [20,50,100],
'max_samples': [0.5,1.0],
'max_features': [0.5,1.0],
'bootstrap': [True, False],
'bootstrap_features': [True, False]}

bagging_regressor_grid = GridSearchCV(BaggingRegressor(random_state=1, n_jobs =- 1), params, cv=3, n_jobs = -1, verbose=1)
bagging_regressor_grid.fit(X_train, y_train)

yu_pred = bagging_regressor_grid.predict(X_test)
print(f'The r2_score of this model gridsearch is :',r2_score(y_test,yu_pred))
```

```python
print(f'The best_score of this model is ',bagging_regressor_grid.best_score_)
print(f'The best_parameter for this model is ',bagging_regressor_grid.best_params_)
```
