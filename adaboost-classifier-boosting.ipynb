```python
# ⚡ AdaBoost (Adaptive Boosting)

# AdaBoost is an Ensemble Learning algorithm

# 🎯 Goal:
# Improve weak models by focusing more
# on previously wrong predictions

# 🧠 Idea:
# Train models sequentially
# where each new model learns from
# mistakes of previous models

# 📦 Weak Learner:
# Usually small Decision Trees (stumps)

# A stump:
# 🌳 Tree with depth = 1

# 🔁 Working of AdaBoost:

# Step 1️⃣
# Train first weak model
# Step 2️⃣
# Check wrong predictions
# Step 3️⃣
# Increase weight of wrongly predicted samples
# Step 4️⃣
# Train next model focusing more on mistakes
# Step 5️⃣
# Combine all weak learners

# 📊 Simple Diagram:

#          📥 Original Dataset
#                   │
#              🌳 Model1
#                   │
#          ❌ Wrong Predictions
#                   │
#        ⬆️ Increase Sample Weights
#                   │
#              🌳 Model2
#                   │
#          ❌ Wrong Predictions
#                   │
#        ⬆️ Increase Sample Weights
#                   │
#              🌳 Model3
#                   │
#         🗳️ Weighted Combination
#                   │
#            ✅ Final Prediction
```

```python
from sklearn.datasets import make_circles
X_c, y_c = make_circles(random_state=42,n_samples=500,)
```

```python
X_c.shape
```

```python
y_c
```

```python
grid = dict()
grid['n_estimators'] = [10, 50, 100, 500]
grid['learning_rate'] = [0.0001, 0.001, 0.01, 0.1, 1.0]
print(grid)

n_estimators = [10, 50, 100, 500]
learning_rate = [0.0001, 0.001, 0.01, 0.1, 1.0]

params = {
'n_estimators': n_estimators,
'learning_rate': learning_rate,
}

print(params)
```

# Applying GridserachCV to find best result on Adaboost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier

from sklearn.model_selection import GridSearchCV


grid_search = GridSearchCV(estimator=AdaBoostClassifier(), param_grid=params, n_jobs = -1, cv=10, scoring='accuracy')
# execute the grid search
grid_result = grid_search.fit(X_c, y_c)
# summarize the best score and configuration
print('Best Score',grid_result.best_score_) # This is cross-validation score
print('Best parameters', grid_result.best_params_)
```

# ⚡ AdaBoost Classifier

 Used for Classification problems

 Final prediction:
 🗳️ Weighted Voting

 Better performing models
 get higher importance/weight


 ⚡ AdaBoost Regressor

 Used for Regression problems

 Final prediction:
 📊 Weighted Average of predictions


 📦 Important Hyperparameters:

 n_estimators
 → Number of weak learners

 learning_rate
 → Contribution of each weak learner

 Small learning_rate:
 → Slower learning
 → More stable
 
 Large learning_rate:
 → Faster learning
 → Risk of overfitting

 estimator
 → Base model used
 Usually:
 DecisionTreeClassifier(max_depth=1)

 random_state
 → Ensures reproducible results
